### Importação dos dados



```python
# Importando as Bibliotecas usadas no Challenge
import pandas as pd
import matplotlib.pyplot as plt

# Importando as bases (Arquivos CSV) do Github da Alura.
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

# Criando os Dataframes de cada loja a partir dos arquivos obtidos
loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)
```

#1. Análise do faturamento

```python
# Comparando o total do faturamento entre as lojas

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
valores = [
    loja1['Preço'].sum(),
    loja2['Preço'].sum(),
    loja3['Preço'].sum(),
    loja4['Preço'].sum()
]

faturamento = pd.Series(valores, index=lojas)

print('Faturamento por Loja:\n')
print(faturamento)
```

```python
# Comparando o faturamento em percentual da Loja 1 (Maior faturamento) com as outras lojas
comparacao = ((faturamento / faturamento['Loja 1']) - 1) * 100

print("Diferença percentual em relação à Loja 1:")
for loja, diff in comparacao.items():
    if loja != 'Loja 1':
        print(f"{loja}: {diff:.2f}%")
```

**A loja 1 teve o maior faturamento e a loja 4 o menor**

# 2. Vendas por Categoria


```python
# Criando Dataframes de produtos por categoria de cada loja
loja1_vendas_cat = loja1.groupby('Categoria do Produto')['Preço'].sum().sort_values(ascending=False)
loja2_vendas_cat = loja2.groupby('Categoria do Produto')['Preço'].sum().sort_values(ascending=False)
loja3_vendas_cat = loja3.groupby('Categoria do Produto')['Preço'].sum().sort_values(ascending=False)
loja4_vendas_cat = loja4.groupby('Categoria do Produto')['Preço'].sum().sort_values(ascending=False)
```

```python
# Concatenando os quatro dataframes para uma analise to total de vendas por produto em cada Loja
vendas_por_categoria = pd.concat([loja1_vendas_cat, loja2_vendas_cat, loja3_vendas_cat, loja4_vendas_cat], axis=1)
vendas_por_categoria.columns = ['Loja1', 'Loja2', 'Loja3', 'Loja4']
vendas_por_categoria
```

**Podemos verificar que o faturamento por produdo e muito proximo e que a Loja 1 tem o maior faturamento nas categorias com maiores valores de faturamento e a loja 4 so superou as outras lojas em categorias com menor faturamento (Brinquedos e Utilidades domesticas)**

# 3. Média de Avaliação das Lojas

```python
# Comparativo da avaliação de compra entre as lojas
avaliacao_media = pd.DataFrame({'Loja1': [loja1['Avaliação da compra'].mean().round(2)],
                                'Loja2': [loja2['Avaliação da compra'].mean().round(2)],
                                'Loja3': [loja3['Avaliação da compra'].mean().round(2)],
                                'Loja4': [loja4['Avaliação da compra'].mean().round(2)]},
                                index=['Avaliação Média'])

avaliacao_media
```

**A avaliação média das lojas tem uma variação muito pequena e foi a unica onde a loja 1 não foi a primeira**

# 4. Produtos Mais e Menos Vendidos

```python
# Criando Dataframe dos produtos mais vendidos por loja
mais_vendidos_loja1 = loja1['Produto'].value_counts().head(1)
mais_vendidos_loja2 = loja2['Produto'].value_counts().head(1)
mais_vendidos_loja3 = loja3['Produto'].value_counts().head(1)
mais_vendidos_loja4 = loja4['Produto'].value_counts().head(1)

# Unindo os Dataframes das lojas para comparação
mais_vendidos = pd.concat([mais_vendidos_loja1, mais_vendidos_loja2, mais_vendidos_loja3, mais_vendidos_loja4], axis=1).fillna(0)
mais_vendidos.columns = ['Loja1', 'Loja2', 'Loja3', 'Loja4']
mais_vendidos
```

```python
# Criando Dataframe dos produtos menos vendidos por loja
menos_vendidos_loja1 = loja1['Produto'].value_counts().tail(1)
menos_vendidos_loja2 = loja2['Produto'].value_counts().tail(1)
menos_vendidos_loja3 = loja3['Produto'].value_counts().tail(1)
menos_vendidos_loja4 = loja4['Produto'].value_counts().tail(1)

# Unindo os Dataframes das lojas para comparação
menos_vendidos = pd.concat([menos_vendidos_loja1, menos_vendidos_loja2, menos_vendidos_loja3, menos_vendidos_loja4], axis=1).fillna(0)
menos_vendidos.columns = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
menos_vendidos
```
